## The Simple Recurrent Network

The networks that we have studied so far have worked by presenting an input and calculating an output based on that input. All of the information that the network has to make a decision comes from the input. In many human tasks, however, the next output depends on the history of the inputs that a person has seen. Perhaps the most obvious examples come from language.

For instance, if we are trying to learn what word should come next in a stream:

The boy who felt sad __ran__ …

The boys who felt sad __run__ …

it is not possible to predict whether it should be __ran__ or __run__ without knowing the history - in this case, whether boy or boys appeared earlier.

To address this problem, Elman (1990) introduced the Simple Recurrent Network (SRN; see Figure 1).



Figure 1: The Simple Recurrent Network

The SRN is a feedforward backpropagation network that maps the input through a hidden layer to the output. To capture prior inputs, the hidden pattern from the previous time step is included as part of the input of the current time step. Because this hidden pattern was also a consequence of the time steps prior to that it captures not just the last time step, but all of the time steps prior to that.



Elman (1990) explored several tasks, but in this tutorial we are going to focus on the prediction of the next word in sentences. To keep the simulation simple, Elman defined a "toy" grammar that displayed some key aspects of language. Words were  assigned to a small number of syntactic/semantic categories as follows:




Then he defined a series of sentence frames that these categories could appear in as follows:



To generate a training corpus, he iteratively chose a sentence frame at random and then sampled words to fill each of the slots. For instance, he might have sampled the NOUN-ANIM VERB-EAT NOUN_FOOD frame first, then he might have sampled the words "cat", "eat", ""cookie" to fill the slots. Then he might have sampled the NOUN-AGRESS VERB-DESTROY NOUN-FRAG frame and sampled the words "dragon", "smash" and "glass". At this point, the corpus would consist of:

cat eat cookie dragon smash glass ...

Note there are no sentence boundary markers, so the network needs to infer where the sentences start and end in order to aid prediction. The following code samples a single sentence. 

```r
noun_human = function (){ sample(c("man", "woman"), 1)}
noun_anim = function (){ sample(c("cat", "mouse"), 1)}
noun_inanim = function (){ sample(c("book", "rock"), 1)}
noun_agress = function (){ sample(c("dragon", "monster"), 1)}
noun_frag = function (){ sample(c("glass", "plate"), 1)}
noun_food = function (){ sample(c("cookie", "cereal"), 1)}
verb_intran = function (){ sample(c("think", "sleep"), 1)}
verb_tran = function (){ sample(c("see", "chase"), 1)}
verb_agpat = function (){ sample(c("move", "break"), 1)}
verb_percept = function (){ sample(c("smell", "see"), 1)}
verb_destroy = function (){ sample(c("break", "smash"), 1)}
verb_eat = function (){ "eat" }

templates = list(c(noun_human, verb_eat, noun_food), 
                 c(noun_human, verb_percept, noun_inanim),
                 c(noun_human, verb_destroy, noun_frag),
                 c(noun_human, verb_intran),
                 c(noun_human, verb_tran, noun_human),
                 c(noun_human, verb_agpat, noun_inanim),
                 c(noun_human, verb_agpat),
                 c(noun_anim, verb_eat, noun_food),
                 c(noun_anim, verb_tran, noun_anim),
                 c(noun_anim, verb_agpat, noun_inanim),
                 c(noun_anim, verb_agpat),
                 c(noun_inanim, verb_agpat),
                 c(noun_agress, verb_destroy, noun_frag),
                 c(noun_agress, verb_eat, noun_human),
                 c(noun_agress, verb_eat, noun_anim),
                 c(noun_agress, verb_eat, noun_food)
                 
                )

getSentence = function(){
    res = c()
    T = length(templates)
    r = sample(1:T, 1)
    template = templates[[r]]
    for (j in 1:length(template)){
        res = c(res, template[[j]]())
    }
    res
}

   
```

And the following code creates a corpus containing multiple samples of sentence sequences. The task of the network is to predict what word comes next. corpus\\$X is the input sequence and corpus\\$Y is the output sequence - that is, the input sequence moved one along.

Note, we have added "zog" to the vocabulary even though it does not appear in the corpus. We will use it later to demonstrate how the network is able to perform with unknown words. Don't worry about it for now, however.

```r
makeCorpus = function(NumberOfSamples = 1, WordsPerSample=10000){
    set.seed(5)
    corpus = NULL
    corpus$NumberOfSamples = NumberOfSamples
    corpus$WordsPerSample = WordsPerSample
    corpus$X = array("#", c(NumberOfSamples, WordsPerSample))
    corpus$Y = array("#", c(NumberOfSamples, WordsPerSample))
    for (i in 1:NumberOfSamples){
        j = 1
        while (j <= WordsPerSample){
            nextSentence = getSentence()
            continue = TRUE
            while ((j <= WordsPerSample+1) & continue){
                nextWord = nextSentence[1]
                if (j <= WordsPerSample){
                    corpus$X[i, j] = nextWord
                }
                if (j > 1){
                    corpus$Y[i, j-1] = nextWord
                }
                j = j + 1
                if (length(nextSentence) == 1){
                    continue = FALSE
                } else {
                    nextSentence = nextSentence[2:length(nextSentence)]
                }
            }
        }
    }
    corpus$vocab = c("man", "woman", "cat", "mouse", "book", "rock", "dragon", "monster",
          "glass", "plate", "cookie", "cereal", "think", "sleep", "chase",
          "move", "break", "smell", "see", "smash", "eat", "zog")
    corpus$V = length(corpus$vocab)
    corpus$I = 1:corpus$V
    names(corpus$I) = corpus$vocab
    corpus
}

corpus = makeCorpus(NumberOfSamples=5, WordsPerSample=10)
corpus
```

Before the corpus can be presented to the network, the words must be converted to vectors. The following code takes a corpus and creates X and Y arrays.

```r
makeArrays = function(corpus){
    result = NULL
    result$X = array(0, c(dim(corpus$X), corpus$V))
    result$Y = array(0, c(dim(corpus$Y), corpus$V))
    
    for (i in 1:corpus$NumberOfSamples){
        for (j in 1:corpus$WordsPerSample){
            result$X[i,j, corpus$I[corpus$X[i,j]]] = 1.0
            result$Y[i,j, corpus$I[corpus$Y[i,j]]] = 1.0
        }
    }
    result
}

arrs = makeArrays(corpus)
```

Now we are ready to train our SRN. We create the corpus, convert it to arrays and then use the trainr method from the rnn package. The number of hidden units is 10 in this example and we will train for 30 epochs (i.e. iterations through the entire training set).  

```r
corpus = makeCorpus(NumberOfSamples=1000, WordsPerSample=20)
arrs = makeArrays(corpus)
model = trainr(arrs$Y, arrs$X, learningrate=0.1, numepochs=30, hidden_dim = 10)
```

Now that we have trained the weights of the network, we can present the input patterns again with the weights fixed to see how the network performs. The following code calculates the average of the output activations of the correct words. Remember that the network is trying to predict what will come next and that is uncertain as the sequences were generated randomly. Therefore, it won't be prefect in it's prediction, but still it will be able to narrow down the options because it is capturing the dependencies.

```r
predictions = predictr(model, arrs$X)
total = array(0, corpus$V)
count = array(0, corpus$V)
for (i in 1:corpus$NumberOfSamples){
    for (j in 1:corpus$WordsPerSample){
        w = corpus$I[corpus$Y[i,j]]
        if (!is.na(w)){
            total[w] = total[w] + predictions[i,j,w]
            count[w] = count[w] + 1
        }
        
    }
}
wordPredictions = signif(total/count, 2)
names(wordPredictions) = corpus$vocab
wordPredictions
```

## Exploring the hidden unit patterns

To reduce the error of prediction, the network must organize the representations in the hidden units. Elman (1990) averaged the patterns corresponding to each word when it was presented and conducted hierarchical cluster analysis on the mean vectors. The following code conducts this analysis for our network.

```r
corpus = makeCorpus(NumberOfSamples=1000, WordsPerSample=20)
arrs = makeArrays(corpus)

plotMeanHidden = function(corpus, arrs, model){
    hidden = predictr(model, arrs$X, hidden=TRUE)[[1]]
    total = array(0, c(corpus$V, dim(hidden)[3]))
    count = array(0, corpus$V)
    for (i in 1:corpus$NumberOfSamples){
        for (j in 1:corpus$WordsPerSample){
            w = corpus$I[corpus$X[i,j]]
            total[w,] = total[w,] + hidden[i,j,]
            count[w] = count[w] + 1  
        }
    }

    result = data.frame(matrix(ncol = dim(total)[2], nrow = 0))
    vocab = NULL
    for (i in 1:corpus$V){
        if (count[i] > 0){ # include only words that were seen in this corpus
            result[nrow(result) + 1,] <- total[i,] / count[i]
            vocab = c(vocab, corpus$vocab[i])
        }
    }
    print (vocab)
    #result = signif(result, 2)
    rownames(result) = vocab
    print(result)
    options(repr.plot.width=8, repr.plot.height=15)
    dist_mat <- dist(result, method = 'euclidean')
    hclust_avg <- hclust(dist_mat, method = 'average')
    plot(as.dendrogram(hclust_avg), horiz=TRUE)
}

plotMeanHidden(corpus, arrs, model)
```

One of the interesting properties of the SRN is that even when a word has not been seen before and as a consequence no training has occurred for that word, the network is able to induce a sensible representation for it. To see that this is the case, we will change the corpus to replace each occurrence of "man" with "zog". Although a node was allocated for "zog" in the network, it didn't appear in the training corpus. We'll then run the input patterns through, calculate the mean of the hidden units for all items in the vocabulary including "zog" and then plot the hierarchical cluster analysis again.

```r
for (i in 1:corpus$NumberOfSamples){
    for (j in 1:corpus$WordsPerSample){
        if (corpus$X[i,j] == "man"){
            corpus$X[i,j] = "zog"
        }
    }
}

arrs = makeArrays(corpus)
plotMeanHidden(corpus, arrs, model)
```
